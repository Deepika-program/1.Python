#BMI Ranges for ADULTS

BMI                   WEIGHT STATUS
Below 18.5             Under Weight
18.5 to 24.9          Healthy Weight
25.0 to 29.9             over Weight
30.0 to 34.9          Obese class 1
35.0 to 39.9          Obese class 2
40.0 and higher        Obese class 3

```python
#Body Mass Index
BMI=float(input("Enter the BMI Index :"))
if(BMI<=18.5):
    print("Under Weight")
elif((BMI>18.5)&(BMI<=24.9)):
    print("Healthy Weight")
elif((BMI>=25.0)&(BMI<=29.9)):
    print("Over Weight")
elif((BMI>=30.0)&(BMI<=34.9)):
    print("Obese Class 1")
elif((BMI>=35.0)&(BMI<=39.9)):
    print("Obese Class 2")
elif(BMI>=40.0):
    print("Obese Class 3")
```
