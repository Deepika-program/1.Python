```python
# Catagory the people by their age like children, adult, citizen, senior citizen...
age=int(input("Enter the age of the people :"))
if(age<18):
    print("Children")
elif(age<35):
    print("Adult")
elif(age<60):
    print("Citizen")
elif(age>60):
    print("Senior citizen")
```
